```python
# Analyze outcome of positional model
#pos_ffnn_trained_on_no_merged_margins
train_data = pd.read_csv("./network_results/train_predictions.csv")
test_data = pd.read_csv("./network_results/test_predictions.csv")
val_data = pd.read_csv("./network_results/val_predictions.csv")

train_labels = train_data["marginal_text"]
train_preds = train_data["preds"]

test_labels = test_data["marginal_text"]
test_preds = test_data["preds"]

val_labels = val_data["marginal_text"]
val_preds = val_data["preds"]


def precision(labels, preds):
  return np.sum((labels == 1.0) & (preds == 1.0)) / np.sum(preds == 1.0)

def recall(labels, preds):
  return np.sum((labels == 1.0) & (preds == 1.0)) / np.sum(labels == 1.0)

def F1(pre, rec):
  return 2/((1/pre)+(1/rec))

def accuracy(labels, preds):
  return np.sum(labels == preds) / len(labels)

def get_metrics(labels, preds):
  acc = accuracy(labels, preds)
  pre = precision(labels, preds)
  rec = recall(labels, preds)
  f_1 = F1(pre, rec)
  return float(acc), float(pre), float(rec), float(f_1)

get_metrics(train_labels,train_preds)
print(train_preds.value_counts(dropna=False))
print(test_preds.value_counts(dropna=False))
print(val_preds.value_counts(dropna=False))

#print("precision",np.sum(train_preds == 1.0))
#print("recall",np.sum(train_labels == 1.0))
print(val_data[val_data.isnull().any(axis=1)].index)
```
