```python
data = pd.read_csv("./data/data_annotated.csv")
train_data = pd.read_csv("./data/train_set.csv")
test_data = pd.read_csv("./data/test_set.csv")
val_data = pd.read_csv("./data/val_set.csv")

print("Absolute number of marginal text")
print("total",sum(data["marginal_text"]))
print("train",sum(train_data["marginal_text"]))
print("val",sum(val_data["marginal_text"]))
print("test",sum(test_data["marginal_text"]))
print("---------")
print("Absolute number of non-marginal text")
print("total",data.shape[0]- sum(data["marginal_text"]))
print("train",train_data.shape[0]-sum(train_data["marginal_text"]))
print("val",val_data.shape[0]-sum(val_data["marginal_text"]))
print("test",test_data.shape[0]-sum(test_data["marginal_text"]))
print("---------")
print("Percetage of marginal text")
print("total",sum(data["marginal_text"])/data.shape[0])
print("train",sum(train_data["marginal_text"])/train_data.shape[0])
print("val",sum(val_data["marginal_text"])/val_data.shape[0])
print("test",sum(test_data["marginal_text"])/test_data.shape[0])
```
